## COUNTING UNIQUE KEYWORDS

```python
from ast import literal_eval
from collections import Counter

result_list = []
for i in range(len(df)):
    result_list.append(literal_eval(df.iloc[i].keywords))
```

```python
keyword_count = Counter(x for xs in result_list for x in set(xs))
```

```python
keyword_count.most_common()
```

## COUNTING LENGTH OF TITLE

```python
from nltk.tokenize import word_tokenize
title_lengths = []
for i in range(len(df)):
    a = df.iloc[i].title
    tokenized = word_tokenize(a)
    title_lengths.append(len(tokenized))
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array(title_lengths)
plt.hist(y);
plt.show()
```

```python
import statistics
print(statistics.stdev(title_lengths))
print(statistics.mean(title_lengths))
```
